```python
# packages
import os
from functools import reduce
from hsfs import Hamiltonian, En_h, h, c, constants_info
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.rcParams['axes.formatter.useoffset'] = False
au_to_khz = 10**-3 * En_h /h
```

# Triplet states, $n=45$

```python
# construct matrix
mat0 = Hamiltonian(n_min=45, n_max=45, S=1, MJ_max=2, l_max=1)
print('number of basis states:', '%d'%mat0.num_states)
mat1 = Hamiltonian(n_min=45, n_max=45, S=1, MJ_max=2, l_max=1)
print('number of basis states:', '%d'%mat1.num_states)
```

```python
# calculate Zeeman map
Bfield = np.linspace(0, 50.0E-6, 101)  # T
zm0 = mat0.zeeman_map(Bfield, Efield=0.0)
zm1 = mat1.zeeman_map(Bfield, Efield=0.0, singlet_triplet_coupling=0.000005)
```

```python
fig, ax = plt.subplots(figsize=(9, 7))

# plot
indexes = range(mat0.num_states)
for ix in indexes:
    ax.plot(Bfield, c*1e-9*zm0[:, ix] / (h * c), ls='-', lw=1.2, alpha=1, c=(0.2, 0.2, 0.8))
indexes = range(mat1.num_states)
for ix in indexes:
    ax.plot(Bfield, c*1e-9*zm1[:, ix] / (h * c), ls='--', lw=1.0, alpha=0.5, c=(0.8, 0.2, 0.0))

# format
ax.set_xlabel('magnetic field (T)')
ax.set_ylabel('energy / $h$ (GHz)')

y_mid_idx = 3
y_mid_E = c*1e-9*zm0[0, y_mid_idx]/ (h * c)
y_range_E = 0.003

#ax.set_xlim(9300, 10000)
ax.set_ylim(y_mid_E-y_range_E, y_mid_E+y_range_E)

# output
plt.grid()
plt.tight_layout()
```
