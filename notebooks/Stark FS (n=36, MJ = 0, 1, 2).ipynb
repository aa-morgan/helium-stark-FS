```python
import os
from functools import reduce
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['axes.formatter.useoffset'] = False
from hsfs import StarkMatrix, h, c
```

```python
nmin = 34
nmax = 38
S = 1
# specify the electric field
field = np.linspace(0, 0.1, 201) # V /cm
# calculate stark maps
stark_matrices = []
stark_maps = []
for i, MJ in enumerate([0, 1, 2]):
    # specify basis
    mat = StarkMatrix(nmin, nmax, S, MJ)
    stark_matrices.append(mat)
    print("MJ = %d. Number of basis states = %d."%(MJ, mat.num_states))
    sm = mat.stark_map(field * 100)
    # diagonalise for each field
    stark_maps.append(sm)
```

```python
# Stark map
fig, ax = plt.subplots(figsize=(6, 4))

for mat, sm in zip(stark_matrices, stark_maps):
    indexs = np.intersect1d(mat.where('n', 36), mat.where('L', 1))
    for ix in indexs:
        ax.plot(1e3*field, 0.01*sm[:, ix] /(h * c), ls='-', lw=1, c='k')

# format
#ax.set_ylim(-84.985, -84.9844)
#ax.set_xlim(0, 300)
ax.grid()
ax.set_xlabel('electric field (mV cm$^{-1}$)')
ax.set_ylabel('energy / $h c$ (cm$^{-1}$)')
# output
plt.tight_layout()
fileout = os.path.join("..", "images", "n36p_fs.png")
plt.savefig(fileout, dpi=200)
plt.show()
```

## $ns \rightarrow np$

```python
# L = 0, |MJ| = 0
mat = stark_matrices[0]
sm = stark_maps[0]
ix_s0 = reduce(np.intersect1d, (mat.where('n', 36),
                                mat.where('L', 0),
                                mat.where('J', 1)))
en_s0 = sm[:, ix_s0[0]]
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

for mat, sm in zip(stark_matrices, stark_maps):
    indexs = np.intersect1d(mat.where('n', 36), mat.where('L', 1))
    for ix in indexs:
        nu = (sm[:, ix] - en_s0) / h
        ax.plot(1e3*field, 1e-9 * nu, c='k', ls='-', lw=1)
        
# format
ax.set_xlabel('electric field (mV cm$^{-1}$)')
ax.set_ylabel(r'$36s \rightarrow 36p$ (GHz)')
plt.grid()
# output
plt.tight_layout()
fileout = os.path.join("..", "images", "36s_36p.png")
plt.savefig(fileout, dpi=200)
plt.show()
```

## $\nu_{02}$ splitting

```python
# L=1, J =0, MJ = 0
mat = stark_matrices[0]
sm = stark_maps[0]
ix_p00 = reduce(np.intersect1d, (mat.where('n', 36),
                                 mat.where('L', 1),
                                 mat.where('J', 0)))
en_p00 = sm[:, ix_p00[0]]
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

for mat, sm in zip(stark_matrices, stark_maps):
    indexs = np.intersect1d(mat.where('n', 36), mat.where('L', 1))
    for ix in indexs:
        nu = (en_p00 - sm[:, ix] ) / h
        ax.plot(1e3*field, 1e-6 * nu, c='k', ls='-', lw=1)

# format
ax.set_ylim(4.55, 4.6)
ax.set_xlabel('electric field (mV cm$^{-1}$)')
ax.set_ylabel(r'$\nu_{0, 2}$ (MHz)')
plt.grid()
# output
plt.tight_layout()
fileout = os.path.join("..", "images", "n36_nu02.png")
plt.savefig(fileout, dpi=200)
plt.show()
```

```python
# L=1, J =1, MJ = 0
mat = stark_matrices[0]
sm = stark_maps[0]
ix_p10 = reduce(np.intersect1d, (mat.where('n', 36),
                                 mat.where('L', 1),
                                 mat.where('J', 1)))
en_p10 = sm[:, ix_p10[0]]
# L=1, J =1, MJ = 1
mat = stark_matrices[1]
sm = stark_maps[1]
ix_p11 = reduce(np.intersect1d, (mat.where('n', 36),
                                 mat.where('L', 1),
                                 mat.where('J', 1)))
en_p11 = sm[:, ix_p11[0]]
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

for mat, sm in zip(stark_matrices, stark_maps):
    indexs = np.intersect1d(mat.where('n', 36), mat.where('L', 1))
    for ix in indexs:
        nu = (en_p10 - sm[:, ix]) / h
        ax.plot(1e3*field, 1e-3 * nu, c='k', ls='-', lw=1)
        nu = (en_p11 - sm[:, ix]) / h
        ax.plot(1e3*field, 1e-3 * nu, c='k', ls='-', lw=1)
        
# format
ax.set_ylim(300, 380)
ax.set_xlabel('electric field (mV cm$^{-1}$)')
ax.set_ylabel(r'$\nu_{1, 2}$ (kHz)')
plt.grid()
# output
plt.tight_layout()
fileout = os.path.join("..", "images", "n36_nu12.png")
plt.savefig(fileout, dpi=200)
plt.show()
```
